## Universality of Pauli Rotations
We can see how some common gates can be decomposed into pauli rotations. There are a few important notes:
1. Due to limits on precision, the resultant states are not exact
2. The $PhaseShfit$ gate can be identical to the $R_Z$ gate when we disregard the *"Global Phase"*, which is not detectable by measurement
