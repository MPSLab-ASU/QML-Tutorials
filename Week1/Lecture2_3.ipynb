## Universality of CNOT+Pauli Rotations
Experiments to decompose any gate into CNOT + Pauli Rotations. Proof can be seen in the paper [Minimal Universal Two-Qubit CNOT-based Circuits](https://arxiv.org/abs/quant-ph/0308033)

```python
draw_and_get_state([qml.CY([0,1])])
print("-"*50)
decompose_and_plot(qml.CY)
```

```python
draw_and_get_state([qml.CH([0,1])])
print("-"*50)
decompose_and_plot(qml.CH)
```
